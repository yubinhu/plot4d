```python
from plot4d import plotter
import numpy as np
from math import cos, sin
import matplotlib.pyplot as plt
import ipympl
%matplotlib widget
```

```python
np.random.seed(41)
f = lambda x, y, z: sin(x)*y*cos(z)-x**3
X = np.random.random(size=(300,3))
plotter.plot4d_data(X, np.array([f(x,y,z) for x,y,z in X]))
plt.show()
```

```python
import numpy as np
from IPython.display import Image

f = lambda x, y, z: sin(x)*y*cos(z)-x**3
z_range = np.linspace(0,2,10)
frame = plotter.Frame2D(xmin=0, xmax=1, ymin=0, ymax=1)
gif_name = plotter.plot4d(f, z_range, frame=frame, func_name='f', path="gif/")

with open(gif_name,'rb') as f:
    display(Image(data=f.read(), format='png'))
```
